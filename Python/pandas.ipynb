```python
fruits_pro={
    'avacado':2.0,
    'guava':2.6,
    'orange':0.9,
    'banana':1.1,
    'apple':0.3,
    'straw':1.0,
    'mango':0.8,
}
```

```python
s2 = pd.Series(fruits_pro,name='protein')
s2
```

```python
#conditinal selection
s2[s2>1]
```

```python
#loical operators and not or

# and = true & true == true
# or = true & false == true any one
# not != 

s2[(s2>0.5) & (s2<2)]
```

```python
s2[(s2>0.5) | (s2<2)]
```

```python
----------Data Cleaning--------
```

```python
df.isnull()
df.dropna()
df.dropna(how='any') # any row that had null values
df.fillna(0)
df['age'].fillna(df['age'].mean()) # filling with mean
df['salary'].fillna(df['salary'].median()) 
```
