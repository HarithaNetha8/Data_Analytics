```python
#pyplot api
#univariate - one variable at a time
```

```python
data= {
    'salary':[10000,30000,4000,54000,230000,50000,39000]
}
```

```python
data
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.shape
```

```python
#line plot :

plt.plot(df['salary'],color='red', marker='1')
plt.grid()
plt.show()
```

```python
#histogram
plt.hist(df['salary'])
plt.show()
```

```python
#boxplot
plt.boxplot(df['salary'])
plt.show()
```

```python
# #univeriate - categorical

df['department'] = ['HR','Sales', 'HR', 'IT', 'Finance', 'Service','IT']
df
```

```python
df.info()
```

```python
count=df['department'].value_counts()
count
```

```python
plt.pie(count,labels=count.index, autopct='%1.1f', explode=[0,0.1,0,0,0])
plt.show()
```

```python
plt.bar(count.index,count)
plt.show()
```

```python
#bivariate analysis - numerical 
```

```python
df['age']=[23,35,29,23,45,22,37]
```

```python
df.head()
```

```python
#scatter pllot :

plt.scatter(df['age'],df['salary'])
plt.show
```

```python
#line plot

plt.plot(df['age'],df['salary'])
plt.show
```

```python
sort_ae=df.sort_values('age')
```

```python
plt.plot(sort_ae['age'],df['salary'])
plt.show
```

```python
#bar cart
plt.bar(sort_ae['age'],df['salary'],color='green')
plt.show
```

```python
#bivarient anlysis - numerical and categorical

hr_sal=df[df['department']=='HR']['salary']
hr_sal
```

```python
IT_sal=df[df['department']=='IT']['salary']
IT_sal
```

```python
sales_sal=df[df['department']=='Sales']['salary']
sales_sal
```

```python
plt.boxplot([hr_sal,IT_sal,sales_sal],labels=['HR','IT','Sales'])
plt.show()
```

```python
salary_by_dept=df.groupby('department')['salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,labels=salary_by_dept.index, autopct='%1.1f',shadow=True, explode=[0,0,0,0,0.1])
plt.show()
```

```python
fig,axs=plt.subplots(1,3, figsize=(10,5))

#line plot

axs[0].plot(sort_ae['age'],df['salary'],color='red', marker='*', linewidth='2',markersize=2)
axs[0].grid()
axs[0].set_title('line plot')
axs[0].set_xlabel('age')
axs[0].set_ylabel('salary')

#histogram
axs[1].hist(df['salary'],bins=5,color='skyblue')
axs[1].set_title('histogram')
axs[1].set_xlabel('salary')
axs[1].set_ylabel('frequency')

#boxplot
axs[2].boxplot(df['salary'])
axs[2].set_title('box plot')
axs[2].set_ylabel('salary')
                
plt.show()
```
