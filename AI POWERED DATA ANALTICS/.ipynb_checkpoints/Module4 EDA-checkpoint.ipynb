# 4.1 Introduction to Exploratory Data Analysis

## What is EDA?
**Exploratory Data Analysis (EDA)** is the process of:
- understanding the dataset
- finding patterns, trends, and relationships
- detecting errors/anomalies

before applying any advanced analysis or modeling.

### Simple Example:

You get a dataset → You don’t directly build a model
First you:
* check missing values
* understand columns
* visualize data
* find patterns

# 4.2 Need for Exploratory Analysis
EDA is important because:
### 1 Understand Data Structure
* What columns exist?
* What type of data?

### 2 Detect Errors Early
* missing values
* wrong entries
* duplicates

### 3 Identify Patterns
Example:
* sales increase during festivals

### 4 Make Better Decisions
EDA helps in:
* choosing correct model
* selecting features

```python
# # CASE STUDY 1: Supermarket Sales Analysis
# ## Step 1: Understand Data
# Columns:

# * product
# * category
# * price
# * quantity
# * date
import pandas as pd
df=pd.read_csv('sales1.csv')
df
```

```python
# ## Step 2: Perform EDA

# ### Check:

# * missing values

df.isnull()
df.isnull().sum()
```

```python
df.fillna(df['CustomerID'].mean())
```

```python
# * duplicates
df.duplicated()
```

```python
df['Revenue']=df['UnitPrice'] *df['Quantity']         
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# ## Step 3: Analyze

# ### Questions:

# * Which product sells most?
product_sell_most= df.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(3)
print(product_sell_most)
```

```python
# Most Sold Product (by Quantity)
top_products_qty = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(5)

print(top_products_qty)
```

```python
# * Peak sales time?
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Day'] = df['InvoiceDate'].dt.day_name()

df.groupby('Day')['Revenue'].sum().sort_values(ascending=False)
```

```python
# ## Step 4: Visualization

# * bar chart → top products

import matplotlib.pyplot as plt

top_products_qty.plot(kind='bar')
plt.title('Most Sold Product (by Quantity)')
plt.show()
```

```python
# * line chart → sales trend

product_sell_most.plot(kind='line')
plt.title('product_sell_most')
plt.show()
```

```python
# ## Step 5: Insights

# Example:

# * BLACK RECORD COVER FRAME has 1627.20 highest sales
# * wednesday as more sales
# * NAMASTE SWAGAT INCENSE sold highhest quantity
```
