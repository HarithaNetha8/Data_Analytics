# Case Study: Sales Data

````python
## Problem Statement:
# Analyze sales data to find:
# * top products
# * sales trends
# * customer insights

## Step 1: Raw Data Issues
import pandas as pd
df=pd.read_csv('sales.csv')
df
````

````python
# * missing values
df.isnull().sum()
````

````python
# * duplicates
df.duplicated()

#sow duplicated data
df[df.duplicated()]
````

````python
# * inconsistent formats
# o wit all uppr or smaller case letters

# ## Step 2: Data Cleaning

# * fill missing values - but not ave missin values in dtaset we can do tese metods and also forward and backward fill
# df.fillna()
# df.fillna().mean()
# df.fillna(0)

# * remove duplicates
df.drop_duplicates()
````

````python
# * fix formats
# fix formate using strin functions and rplace function

# ## Step 3: Transformation
# * convert date format
df['date']=pd.to_datetime(df['date'])
````

````python
# # * create new columns (profit, revenue)
df['total_amount']=df['price']*df['quantity']
df
````

````python

# ## Step 4: Analysis

# * total sales
t_s=df['total_amount'].sum()
print('TOTALSALES :',t_s)

# * top-selling products 
top_sell=df.groupby('product')['total_amount'].sum().sort_values(ascending=False)
print('TOP_SELLIN_PRODUCT',top_sell)

# * monthly trends
df['date']=pd.to_datetime(df['date'])
monthly_s=df.groupby(df['date'].dt.to_period('M'))['total_amount'].sum()
print(monthly_s)
````

````python
# ## Step 5: Visualization

import matplotlib.pyplot as plt
````

````python
top_sell.plot(kind='bar')
plt.title('top sellin product')
plt.show()
````

````python
monthly_s.plot(kind='bar')
plt.title('montly sales by revenue')
plt.show()
````

````python
# ## Step 6: Insights

# * laptop has highest sales
# * 2024-3 having highest revenue by month-march
````
