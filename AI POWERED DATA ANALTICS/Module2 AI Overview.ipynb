# 2.1 Introduction to AI

##What is Artificial Intelligence?

**Artificial Intelligence (AI)** is the ability of machines to **simulate human intelligence** such as:

* learning
* reasoning
* problem-solving
* decision-making

### Simple Example:

* Netflix recommending movies
* Google Maps suggesting routes
* Chatbots answering queries

# 2.2 Brief History and Evolution of AI

- 1950s – Birth of AI
- 1960s–70s – Early AI Systems
- 1980s – Expert Systems
- 2000s – Machine Learning Growth
- 2010s – Deep Learning Revolution
- 2020s – Generative AI Era

# 2.4 AI vs ML vs Deep Learning vs GenAI

| Term          | Meaning                                         |
| ------------- | ----------------------------------------------- |
| AI            | Broad concept (machines behaving intelligently) |
| ML            | Subset of AI (learning from data)               |
| Deep Learning | Subset of ML using neural networks              |
| GenAI         | AI that creates new content                     |


# 2.5 What is Machine Learning
Machine Learning (ML) is a subset of AI where machines **learn from data without explicit programming**.

### Example:

Spam detection:

* learns from past emails
* predicts new spam

### Types of ML

1. Supervised Learning
2. Unsupervised Learning
3. Reinforcement Learning
   
# 2.6 Basics of NLP (Natural Language Processing)

NLP allows machines to **understand human language**.

### Examples:

* Chatbots
* Voice assistants
* Translation
  
### NLP Tasks:

* Text classification
* Sentiment analysis
* Named entity recognition
* Language translation
  
# 2.7 What is Deep Learning

Deep Learning is a subset of ML using **multi-layer neural networks**.

### Used in:

* Image recognition
* Speech recognition
* Self-driving cars
  
# 2.8 Introduction to Neural Networks

Neural Networks are inspired by the **human brain**.

### Structure:

* Input layer
* Hidden layers
* Output layer

Each node = neuron

### Example:

Input → Process → Output

Image → Neural Network → Identify object

# 2.9 Why Deep Learning is Critical in GenAI

Generative AI relies on **deep learning models** because:

* handles large data
* understands patterns
* generates human-like content

Without deep learning → GenAI is not possible.

# 2.12 What are LLMs?

LLMs (Large Language Models) are AI models trained on **huge text datasets**.

### Examples:

* ChatGPT, Gemini

# 2.13 How LLMs are Trained

### Step 1: Data Collection

Massive text data (books, websites)

### Step 2: Training

Model learns patterns in language

### Step 3: Fine-tuning

Improved using human feedback

### Step 4: Deployment

Used in apps like chatbots

# 2.14 Prompt Engineering Fundamentals

Prompt engineering = **how you ask AI matters**

### Good Prompt Example:

Bad:
“Analyze data”

Good:
“Analyze sales data and identify top 3 trends with insights”

### Tips:

* be specific
* give context
* define output format
