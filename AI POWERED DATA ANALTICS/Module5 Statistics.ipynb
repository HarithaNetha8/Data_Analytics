# 5.2 Types of Statistics

Just **summarizes data** works on past data 

Example:

* average salary

## 2. Inferential Statistics (What will happen?)
 Uses sample to **predict or conclude**

Example:

# 5.3 Population vs Sample

## Population

Entire data(overall data present)
Example:
All customers of Amazon

## Sample

Small part of data(some part of data)

Example:
1000 customers outof all

### Why use sample?

* faster
* cheaper
* still accurate

#  5.4 Descriptive Statistics (Core Concepts)
## 1. Central Tendency
### Mean (Average)
Sum / Total values
Example:
(10 + 20 + 30) / 3 = 20
<!-- df['Revenue'].mean() -->

### Median
Middle value
Example:
10, 20, 30  → 20

### Mode
Most frequent
Example:
10, 10, 20  → 10


## 2. Variance(spread of data) & Standard Deviation(sqrt of variance)
Measure how **spread out** data is
Variance → spread
Std Dev → variation
### Example:
Dataset A:
10, 10, 10 → low variation

Dataset B:
5, 20, 40 → high variation

 **Standard deviation = spread of data**

##  3. Distribution
How data is spread
### Normal Distribution (Ideal)
* symmetric
* bell-shaped

### Skewed Distribution
* left skew
* right skew

Example:
Income → right skew (few people earn more)

# 5.5 Correlation vs Covariance
## Correlation
Measures **relationship strength**

Range: -1 to +1
- +1  -> strong positive
- -1  -> strong negative
- 0  ->  no relation     

### Example:
* more study → more marks
positive correlation

## Covariance
Shows direction (but not strength clearly)
Measures direction of relationship.
Less interpretable than correlation.

* Correlation = **how strong**
* Covariance = **how direction**

# 5.6 Probability & Random Variables

## Probability
Chance of something happening or
Shows likelihood of outcomes.
Example:
* coin → 50% heads

## Random Variable
A variable with uncertain values or
Value that changes randomly
Example:
* number of customers per day

## Probability Distribution
Shows how values are spread

### Types:

#### Discrete
* countable values (1,2,3) - dice roll

#### Continuous
* measurable (height, weight)
Common Distributions: Normal distribution, Binomial distribution

# 5.7 Inferential Statistics (Decision Making)

## Hypothesis Testing
Used to test assumptions
Steps:

1 Null Hypothesis (H₀)
→ No effect

2 Alternative Hypothesis (H₁)
→ Effect exists

Statistical Significance Measured using
3 p-value
Decision rule:

* p < 0.05 → reject H₀
* p > 0.05 → accept H₀
* 
### Example:
Did marketing campaign increase sales?
Simple meaning:
“Is result real or just luck?”

# 5.8 Type I & Type II Errors

## Type I Error (False Positive)
You say something works but it doesn’t

Rejecting true null hypothesis.

Example:
Saying campaign worked when it didn’t

## Type II Error (False Negative)
You say something doesn’t work but it does

Failing to reject false null hypothesis.

Example:
Saying campaign didn’t work when it did

 Type I  -> False alarm       
 Type II -> Missed opportunity

Case Study 2: Email Campaign

```python
import pandas as pd

df = pd.read_csv('email_campaign_dataset.csv')
df
```

```python
df['difference'] = df['after_sales'] - df['before_sales']
df['difference'].mean()
```

```python
from scipy.stats import ttest_rel

t_stat, p_value = ttest_rel(df['before_sales'], df['after_sales'])

print("P-value:", p_value)
```

```python
df[['before_sales','after_sales']].mean().plot(kind='bar')
```
