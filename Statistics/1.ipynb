**Median (Middle Value)**
🔹 Definition : Middle value when data is sorted
🔹 Steps : Sort data -> Odd count : middle value -> Even count:average of two middle values
🔹 Example
Data: 10, 20, 30 → Median = 20
Data: 10, 20, 30, 100 → Median = (20+30)/2 = 25

**Mode (Most Frequent Value)**
Definition:Value that appears most often
🔹 Example
Data: 2, 4, 4, 6 → Mode = 4
🔹 Types
Unimodal → one mode
Bimodal → two modes
No mode → all values unique
